# Dataset Overview

Answer:

- Number of samples
- Number of classes
- Names of the classes


```python
from datasets import load_dataset

dataset = load_dataset("dair-ai/emotion")

num_samples = sum(split.num_rows for split in dataset.values())
label_names = dataset["train"].features["label"].names
num_classes = len(label_names)

print("Splits:", dataset.keys())
print("Number of samples:", num_samples)
print("Number of classes:", num_classes)
print("Class names:", label_names)
```

# Class Distribution

Generate a chart showing:

- joy
- sadness
- anger
- fear
- love
- surprise

```python
import pandas as pd
import matplotlib.pyplot as plt

train_df = dataset["train"].to_pandas()
train_df["emotion"] = train_df["label"].apply(lambda x: label_names[x])

counts = train_df["emotion"].value_counts().reindex(label_names)

counts.plot(kind="bar")
plt.title("Class Distribution")
plt.xlabel("Emotion")
plt.ylabel(("Number of examples"))
plt.show()
```

# Text Length Analysis

Calculate:

- smallest text
- largest text
- average number of words

```python
train_df["word_count"] = train_df["text"].apply(lambda text: len(text.split()))

print("Smallest text length:", train_df["word_count"].min())
print("Largest text length:", train_df["word_count"].max())
print("Average words:", train_df["word_count"].mean())
```
